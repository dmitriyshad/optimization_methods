### Задание 1

Базовая модель, используемая в Causal Impact, следующая:

$$ y_t = Z^T_t \alpha_t + \varepsilon_t, \\
    \alpha_{t+1} = T_t \alpha_t + R_t \eta_t. $$
    
Первое уравнение называется "observation equation" (уравнение наблюдения), второе - "state equation" (уравнение состояния). Если на пальцах, то временной ряд прогнозируется в два этапа - сначала из уравнения состояния определяется новое состояние по предыдущему + некоторый случайный шум, затем определяется новое значение ряда по состоянию + опять же некоторый новый шум, не зависящий от предыдущего. Состояние $\alpha_t$ это вектор некоторой фиксированной размерности, $T_t, R_t$ - матрицы соответствующих размеров, $Z_t$ - вектор. Интересный нюанс кроется в размере $\eta_t$ - он может быть меньше размера вектора состояния; такая нотация позволяет учесть ситуации, когда случайный шум добавляется не к каждой компоненте независимо. Как описано в статье Бродерсена, этот нюанс критичен в модели для сезональности.

Как я понял из статьи, в вектор состояния зашиты локальный линейный тренд и сезональность. Так что это даже не особо абстрактная модель. Но самая главная часть состояния - регрессионная часть, которая позволяет её оценить, используя информацию с похожих данных, но в которых исследуемого воздействия не было. Отсюда вытекают границы применимости модели - эффект, исследуемый для временного ряда не должен влиять на другие ряды, в которых эффект отсутствует. В статье приводится пример рекламной кампании в США, который мог докатиться до рынков Канады или Великобритании.

```r
library(CausalImpact)
```

Пройдемся по параметрам CausalImpact. Для начала откроем документацию:

```r
?CausalImpact::CausalImpact
```

* data - собственно, данные.
* pre.period - интервал данных до взаимодействия.
* post.period - интервал данных после взаимодействия.
* model.args - аргументы для обучения модели, как раз то, что нам интересно.
* bsts.model - вместо того, чтобы CausualImpact сам настраивал модель, можно передать уже обученную модель. В этом случае не нужны параметры data, pre.period, post.period, а модель нужно обучить на данных, где значения после воздействия заменены на NA.
* post.period.response - значения после воздействия, указывается только при указании bsts.model.
* alpha - уровень значимости.

Пройдемся теперь по возможным аргументам для обучения модели:
* niter - число итераций при использовании Markov chain Monte Carlo. Довольно простой параметр - чем он больше, тем точность выше. Ничего особенного.
* standardize.data - булевый параметр, указывающий, нужно ли "станартизировать столбцы данных". Как я понял, если выставить этот параметр FALSE, то это будет эквивалентно равномерному априорному распределению на множестве параметров. Тоже ничего сверхестественного, лучше ставить TRUE.
* prior.level.sd - стандартное отклонение шума в обучаемой модели. Естественно выставлять это значение поменьше, если волотильность ряда невысокая.
* nseasons - период сезональной компоненты. В CausualImpact поддерживается только одна сезональная компонента, если нужно больше - придется использовать bsts.
* season.duration - число наблюдений в единице периода. Позволяет учитывать, например, недельную сезональность, если есть ежечасные данные.
* dynamic.regression - если FALSE, то параметры модели не зависят от времени ($Z_t, R_t, T_t$, дисперсии шумов на обоих шагах). Иногда есть смысл выставить TRUE.

Побалуемся немного с параметрами: попробуем разные значения prior.level.sd.

```r
set.seed(1)
x1 <- 100 + arima.sim(model = list(1,2,0), n = 100)
y <- 1.2 * x1
noise <- c(rnorm(70, sd = 0.1), rnorm(30, sd = 1))
y <- y + noise
y[71:100] <- y[71:100] + 0.5
data <- cbind(y, x1)
matplot(data, type = "l")
```

```r
pre.period <- c(1, 70)
post.period <- c(71, 100)
impact <- CausalImpact(data, pre.period, post.period)
plot(impact)
```

```r
pre.period <- c(1, 70)
post.period <- c(71, 100)
impact <- CausalImpact(data, pre.period, post.period, model.args = list(prior.level.sd = 0.1))
plot(impact)
```

Первый эксперимент проводился со стандартным значением prior.level.sd, второй - с "реальным". Как видно из результатов, оба эксперимента дают схожий прогноз эффекта, но разные доверительные интервалы.

Применение динамической модели в данном синтетическом примере сильно ухудшает результаты:

```r
pre.period <- c(1, 70)
post.period <- c(71, 100)
impact <- CausalImpact(data, pre.period, post.period, model.args = list(dynamic.regression = TRUE))
plot(impact)
```

Отказ от стандартизации данных также приводит к расширению доверительных интервалов:

```r
pre.period <- c(1, 70)
post.period <- c(71, 100)
impact <- CausalImpact(data, pre.period, post.period, model.args = list(standardize.data = FALSE))
plot(impact)
```

В качестве таргета возьмем среднее число запросов по всем запросам, а в качестве предикторов - среднее число запросов по запросам, не связанным с картинками. В нулевом приближении кажется, что людям, не ищущим картинки, будет по барабану на то, что картинки стали искаться лучше, поэтому эффекта в этом ряду наблюдаться не должно. 

Проблемы, с которыми мы точно столкнемся: поисковик Яндекса, как все нормальные поисковики, включает в поисковую выдачу картинки, причем очень часто они находятся в самом верху, привлекая внимание пользователя. Для того, чтобы исключить такой эффект, в первом приближении можно попробовать убрать из предиктора запросы, когда юзер в первую очередь кликнул на картинку, а не на результат обычной поисковой выдачи.
